```python
%load_ext autoreload
%autoreload 2

import torch
import chai

import json
import os

import numpy as np
```

```python


data = []
for num_nodes in range(40):
    for num_images in [1,5,10,50,100,500,1000,5000,10000]:
        with open(f'../server_timedata/ml_test_{num_nodes + 1}_{num_images}.json') as f:
            d = json.load(f)
            median = np.median([float(i) for i in d['times']])
            data_row = {
                'num_nodes': d['num_nodes'],
                'num_images': d['num_images'],
                'time': float(d['num_images']) / float(d['avg']),
                'stddev': float(d['num_images']) / d['stddev'],
                'median': float(d['num_images']) / median
            }
            data.append(data_row)
```

```python
import csv
keys = data[0].keys()

with open('time_data.csv','w') as f:
    dict_writer = csv.DictWriter(f,keys)
    dict_writer.writeheader()
    dict_writer.writerows(data)
```
